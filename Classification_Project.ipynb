<p style="background-color:#34495E; font-family:Georgia, serif; color:#FDFEFE; font-size:150%; text-align:center; border-radius:12px; padding:10px;">
 Feature Engineering: Outlier Detection using Z-Score
</p>

<!-- Section Description -->
<p style="font-family:Verdana, sans-serif; font-size:105%; color:#2C3E50;">
Outliers can negatively impact model performance by skewing feature distributions and affecting model assumptions. Here, we detect and address outliers using the Z-score method to enhance data quality and robustness.
</p>

<!-- Outlier Treatment Steps -->
<ul style="line-height:1.8; color:#34495E; font-family:Verdana, sans-serif; font-size:105%;">
  <li> Calculate Z-scores for numerical features such as  <b>LoanAmount</b></li>
  <li> Mark values as outliers if their absolute Z-score exceeds a threshold (typically 3.0)</li>
  <li> Remove or cap the identified outliers to reduce skewness and improve model generalization</li>
</ul>

<!-- Summary -->
<p style="font-family:Verdana, sans-serif; font-size:105%; color:#2C3E50;">
Using Z-score for outlier detection ensures that extreme anomalies are addressed without arbitrarily trimming valid data, leading to a cleaner and more reliable dataset for modeling.
</p>


```python
len_1 = len(df)

from scipy.stats import zscore
df['z_score'] = zscore(df['LoanAmount'])
df = df[df['z_score'].abs() <=3]  # keep rows within 3 standard deviations
df.drop(columns='z_score', inplace=True)

len_2 = len(df)
print("Percentage of data removed in outlier detection:", round(((len_1 - len_2) * 100) / len_1, 2), "%")
```

```python
# --- Feature Engineering ---
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']
df['Log_TotalIncome'] = np.log1p(df['TotalIncome'])
df['Log_LoanAmount'] = np.log1p(df['LoanAmount'])

# EMI = LoanAmount / Loan_Term
df['LoanTerm_Years'] = df['Loan_Amount_Term'] / 12
df['EMI'] = df['LoanAmount'] / df['LoanTerm_Years']

# Income-to-loan ratio
df['Income_Loan_Ratio'] = df['TotalIncome'] / df['LoanAmount']

# Coapplicant flag
df['Has_Coapplicant'] = df['CoapplicantIncome'].apply(lambda x: 1 if x > 0 else 0)

# Clean 'Dependents'
df['Dependents'] = df['Dependents'].replace('3+', 3).astype(int)

# Binary self-employed
df['Is_Self_Employed'] = df['Self_Employed'].map({'Yes': 1, 'No': 0})

# Drop unused
df.drop(['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term','Loan_ID','Self_Employed','TotalIncome'], axis=1, inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression

def classify(model, X, y):
    # Train-test split with stratification
    X_train_resampled, X_test, y_train_resampled, y_test = train_test_split(
        X, y, stratify=y, test_size=0.25, random_state=42
    )
    
    # Scale the resampled train and original test set
    scaler = MinMaxScaler()
    X_train_resampled = scaler.fit_transform(X_train_resampled)
    X_test = scaler.transform(X_test)
    
    # Train the model
    model.fit(X_train_resampled, y_train_resampled)
    
    # Evaluate on test set
    y_pred = model.predict(X_test)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print(f"üîπ Accuracy:  {acc:.4f}")
    print(f"üîπ Precision: {prec:.4f}")
    print(f"üîπ Recall:    {rec:.4f}")
    print(f"üîπ F1 Score:  {f1:.4f}")
 
    
    # Cross-validation with scaling
    pipeline = make_pipeline(MinMaxScaler(), model)
    scores = cross_val_score(pipeline, X, y, cv=5)
    print("üî∏ Cross-validation Accuracy:", np.mean(scores) * 100)
    
    print("\nüîπ Confusion Matrix:\n", cm)

    # Plot confusion matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    return model
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
from scipy.stats import loguniform
import seaborn as sns
import matplotlib.pyplot as plt

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE to training data only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Scale the features
scaler = MinMaxScaler()
X_train_resampled = scaler.fit_transform(X_train_resampled)
X_test_scaled = scaler.transform(X_test)

# Define the model
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Define hyperparameter search space
param_dist_lr = {
    'C': loguniform(0.001,0.5,1,10),
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'],
    'class_weight': [None, 'balanced']
}

# Randomized search
random_search_lr = RandomizedSearchCV(
    estimator=logreg,
    param_distributions=param_dist_lr,
    n_iter=30,
    scoring='f1',
    cv=5,
    random_state=42,
    n_jobs=-1
)

# Fit model
random_search_lr.fit(X_train_resampled, y_train_resampled)
best_lr = random_search_lr.best_estimator_

# Evaluation function
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print(f"üìò {model_name} Evaluation:")
    print(f"üîπ Accuracy:  {acc:.4f}")
    print(f"üîπ Precision: {prec:.4f}")
    print(f"üîπ Recall:    {rec:.4f}")
    print(f"üîπ F1 Score:  {f1:.4f}")
    print("\nüîπ Confusion Matrix:\n", cm)

    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# ‚úÖ Evaluate best logistic regression model
evaluate_model(best_lr, X_test_scaled, y_test, "Logistic Regression")
```

Voting Classifier with Soft Voting

```python
from sklearn.ensemble import VotingClassifier, RandomForestClassifier, GradientBoostingClassifier 
from sklearn.linear_model import  LogisticRegression
from catboost import CatBoostClassifier
from xgboost import XGBClassifier 
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report, confusion_matrix

# Define base models
model1 = LogisticRegression(max_iter=1000, random_state=42)
model2 = RandomForestClassifier(n_estimators=100, random_state=42)
model3 = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model4 = CatBoostClassifier(iterations=100, random_state=42)
model5 = GradientBoostingClassifier(random_state=42)


# Create the Voting Classifier with soft voting
voting_model = VotingClassifier(
    estimators=[
        ('Logistic', model1),
        ('RandomForest', model2),
        ('XGBoost', model3),
        ('CatBoostClassifier',model4),
        ('GradientBoostingClassifier', model5)
    ],
    voting='soft'
)

# Fit the ensemble model
voting_model.fit(X_train_resampled, y_train_resampled)

# Predictions
y_pred = voting_model.predict(X_test)

# Evaluate performance
print("Voting Classifier Results:\n")
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)


print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```
